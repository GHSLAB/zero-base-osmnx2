```python
import osmnx as ox

G = ox.load_graphml(filepath="./data/qz_network_all.graphml")
# 可视化街道网络
fig, ax = ox.plot_graph(G)
```

```python
# 图graph 转换至geopandas GeoDataFrames node节点 edge边
gdf_nodes, gdf_edges = ox.graph_to_gdfs(G)
gdf_nodes.head()
```
